# Расписание автобуса

```python
def generate_bus_schedule(start_time, end_time, interval_minutes):
    """
    Генерирует расписание автобусов.

    Args:
        start_time (str): Время начала движения в формате HH:MM.
        end_time (str): Время окончания движения в формате HH:MM.
        interval_minutes (int): Интервал движения автобусов в минутах.

    Returns:
        list: Список строк, представляющих время отправления автобусов.
    """
    schedule = []
    current_hour, current_minute = map(int, start_time.split(':'))
    end_hour, end_minute = map(int, end_time.split(':'))

    # Преобразуем время в минуты для удобства расчетов
    start_total_minutes = current_hour * 60 + current_minute
    end_total_minutes = end_hour * 60 + end_minute

    current_total_minutes = start_total_minutes

    while current_total_minutes <= end_total_minutes:
        hour = current_total_minutes // 60
        minute = current_total_minutes % 60
        schedule.append(f"{hour:02}:{minute:02}")
        current_total_minutes += interval_minutes

    return schedule

# Пример использования:
start = "06:00"
end = "22:00"
interval = 30  # Автобусы ходят каждые 30 минут

bus_schedule = generate_bus_schedule(start, end, interval)

print("Расписание автобусов:")
for time in bus_schedule:
    print(time)
```
